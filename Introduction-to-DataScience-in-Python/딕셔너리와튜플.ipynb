**딕셔너리 만들기**

딕셔너리는 여러 키와 값을 저장해두고

필요할때 값을 꺼내 쓸 수 있는기능이다.

```
wintable = {
    '가위':'보',
    '바위':'가위',
    '보':'바위'
}
print(wintable['가위'])
```

윈테이블은 각 가위, 바위, 보의 승리조건을 담은 딕셔너리이다.

대괄호부분은 리스트에서 값을 가져오는 문법처럼

키를 이용하여 값을 꺼내서 사용한다.

이름과 값은 콜론으로 구분하고 여러 키, 값은 콜론으로 구문한다.

```
def rsp(mine,yours):
  if mine == yours:
    return 'draw'
  elif wintable[mine] == yours:
    return 'win'
  else:
    return 'lose'

result = rsp('가위','바위')
print(result)
```

```
messages = {
    'win':'이겼다!',
    'draw':'비겼네.',
    'lose':'졌어...'
}
print(messages[result])
```

**딕셔너리와 반복문**



```
ages = {'Tod':35,'Jane':23,'Paul':62}

print(ages)
```

내가 원하는 양식으로 프린트 해보기

```
for key in ages.keys():
  print(key)

for key in ages: #키 생략 가능
  print(key)

for value in ages.values():
  print(value)
```

```
for key in ages.keys():
  print('{}의 나이는 {}입니다.'.format(key,ages[key]))

for key,value in ages.items():
  print('{}의 나이는 {}입니다.'.format(key,value))
```

**리스트와 딕셔너리 비교**

리스트와다르게 딕셔너리는 순서가 없다.

찾기

```
dict = {'one':1, 'two':2}

print('one' in dict.keys())

print(1 in dict.values())

print('one' in dict)
print(1 in dict)
```

결합

```
dict1 = {1:100, 2:200}
dict2 = {1:1000,3:300}

dict1.update(dict2)
print(dict1)

dict1 = {1:100, 2:200}
dict2={1:1000,3:300}

dict2.update(dict1)
print(dict2)
```

튜플은 변수간의 값을 바꾸거나 

여러값을 한번에 전달할때 유용하게 사용할 수 있다.

**튜플을이용한 packing, unpacking**
파이썬은 튜플을 이용해 하나의변수를 여러 변수에 대입할 수 있다.

```
a, b = 1, 2
c = 3, 4
d, e = c # c의 값을 언패킹하여 d, e에 값을 넣었다.
f = d, e # 변수 d와 e를 f에 패킹

print(a,b,c,d,e,f)
```

이것을 활용하면 두 변수의 값을 바꿀 때 임시변수가 필요없고

함수의 리턴 값으로 여러 값을 전달할 수 있다.

```
#두 변수의 값을 바꾸기

#일반적인 방법
x = 5
y = 10
temp = x
x = y
y = temp
print(x,y)


#튜플을 활용한 방법
x = 5
y = 10
x,y = y,x
print(x,y)
```

```
# 함수의 리턴값으로 여러 값 전달하기
def tuple_func():
  return 1,2

  q,w = tuple_func()

  print(q,w)
```
